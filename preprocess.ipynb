```
import pandas as pd
import numpy as np
file_name = "everytime_text" #@param {type:"string"}
df = pd.read_csv(file_name+".csv") 
label_dict = {
    'Spoiled': 0,
    'Default': 1
}
df = df.reindex(np.random.permutation(df.index))

df
```

```
text_data = open('{}.txt'.format(file_name), 'w')
```

```
import random
import re
import copy
for idx, item in df.iterrows():
  text = item['text']
  if text == '':
    print('no sent')
    continue
  orig_text = copy.deepcopy(text)
  ## <unused0> => <url>, <unused1> => <link>, <unused2> => \n
  try:
    text = text.replace('<br />','<unused2>')
  except AttributeError:
    print(text)
    continue
  if len(text.split('\t')) > 1:
    print(len(text.split('\t'))) 
  text = re.sub("""(?:(?:https?|ftp):\/\/|\b(?:[a-z\d]+\.))(?:(?:[^\s()<>]+|\((?:[^\s()<>]+|(?:\([^\s()<>]+\)))?\))+(?:\((?:[^\s()<>]+|(?:\(?:[^\s()<>]+\)))?\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))?""","<unused0>", text)
  text = re.sub('<\s*a[^>]*>(.*?)<\s*/\s*a>', "<unused1>", text)
  
  write_string = '{}\t{}\n'.format(file_name, text)
  text_data.write(write_string) # if ratio > test_file_rate else test_data.write(write_string)
text_data.close()
```

```
import random
import re
import copy
import kss

for idx, item in df.iterrows():
  text = item['text']
  if text == '':
    continue
  orig_text = copy.deepcopy(text)
  ## <unused0> => <url>, <unused1> => <link>, <unused2> => \n
  try:
    text = re.sub("""(?:(?:https?|ftp):\/\/|\b(?:[a-z\d]+\.))(?:(?:[^\s()<>]+|\((?:[^\s()<>]+|(?:\([^\s()<>]+\)))?\))+(?:\((?:[^\s()<>]+|(?:\(?:[^\s()<>]+\)))?\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))?""","<unused0>", text)
  except AttributeError:
    print(text)
    continue
  text = re.sub('<\s*a[^>]*>(.*?)<\s*/\s*a>', "<unused1>", text)
  save_text=''
  for sent in kss.split_sentences(text):
    carrige_return = '<unused2>' if save_text is not '' else ''
    # print(carrige_return)
    save_text += carrige_return
    save_text += sent
  
  write_string = '{}\t{}\n'.format(file_name, save_text)
  text_data.write(write_string) # if ratio > test_file_rate else test_data.write(write_string)
text_data.close()
```
